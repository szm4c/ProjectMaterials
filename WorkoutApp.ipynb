```python
import requests as req
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
# informacja, ktora zczytalem ze strony, keys to tagi ze strony, values to ilosc stron do zczytania przez program
tags = {
    'abs': 8,
    'arms': 5,
    'back': 5,
    'butt-hips': 9,
    'chest': 4,
    'full-body-integrated': 7,
    'legs-calves-and-shins': 3,
    'neck': 1,
    'shoulders': 6,
    'legs-thighs': 8
}

# koncowa lista cwiczen
exercises = list()

for tag, val in tags.items():
    num_of_pages = val
    for i in range(1, num_of_pages + 1):
        page = req.get(f'https://www.acefitness.org/education-and-resources/lifestyle/exercise-library/body-part/{tag}/?page={i}')
        if str(page.status_code).startswith('2'):
            soup = BeautifulSoup(page.content, 'html.parser')
            
            # zczytuje nazwy i inne parametry cwiczen
            exercises_headers = soup.find_all('h2' , class_='exercise-card__title')

            for exercise in exercises_headers:
                exercises.append(exercise.get_text())
    
    else:
        print('Strona nie odpowiada. "page.status_code" nie zaczyna si "2"')
        
```
